```python
import pandas as pd
```

```python
df = pd.read_html('https://fbref.com/en/squads/822bd0ba/Liverpool-Stats', attrs={'id': 'stats_standard_9'})[0]
```

```python
df.head()
```

```python
# You'll also notice that the columns are multi-indexed, so we'll flatten the columns
df.columns = ['_'.join(col).strip() for col in df.columns.values]
```

```python
# You can remove some of the names if you want
df.columns = df.columns.str.replace('Unnamed: 0_level_0_', '')
```

```python
df.columns.values
```

```python
df = df.rename(columns={
    'Unnamed: 1_level_0_Nation': 'Nation',
    'Unnamed: 2_level_0_Comp': 'Comp',
    'Unnamed: 2_level_0_Pos': 'Pos',
    'Unnamed: 3_level_0_Age': 'Age',
    'Unnamed: 4_level_0_MP': 'MP',
})
df.columns.values

```
